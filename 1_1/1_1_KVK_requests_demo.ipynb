# GET REQUEST: bitcoin ár lekérés Kucoin API-n keresztül

```python
import requests

response = requests.get("https://api.kucoin.com/api/v1/market/orderbook/level1?symbol=BTC-USDT")

btc_price = response.json()["data"]["price"]

btc_price
```

# POST REQUEST: openAI API hívás és "beszélgetés"
